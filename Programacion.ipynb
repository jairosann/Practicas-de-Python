# Resumen de cada columna

```python
def resumen_columna(df, col):
    serie = df[col]

    print(f"=== Resumen de '{col}' ===")
    print("Tipo:", serie.dtype)
    print("N nulos:", serie.isna().sum())
    
    # Si es numérica, imprime estadísticas numéricas
    if pd.api.types.is_numeric_dtype(serie):
        print("Media:", serie.mean())
        print("Mediana:", serie.median())
        print("Moda:", list(serie.mode()))
        print("Desviación típica:", serie.std())
        print("Mínimo:", serie.min())
        print("Máximo:", serie.max())
        print("Percentiles (25-50-75):")
        print(serie.quantile([0.25, 0.5, 0.75]))
    else:
        # Si es categórica o string
        print("Valores únicos:", serie.nunique())
        print("Moda:", list(serie.mode()))
        print("Frecuencias:")
        print(serie.value_counts())


columna = 'Gender'
df = pd.read_csv('StudentPerformanceFactors.csv')
resumen_columna(df, columna)
```

## Mostrar las graficas de columnas con NaN

```python
import matplotlib.pyplot as plt

# 1. Teacher_Quality
student_performance['Teacher_Quality'].value_counts(dropna=False).plot(kind='bar')
plt.title('Teacher_Quality')
plt.show()
print(student_performance['Teacher_Quality'].value_counts(dropna=False))
```

```python
# 2. Parental_Education_Level
student_performance['Parental_Education_Level'].value_counts(dropna=False).plot(kind='bar')
plt.title('Parental_Education_Level')
plt.show()

print(student_performance['Parental_Education_Level'].value_counts(dropna=False))
```

```python
# 3. Distance_from_Home
student_performance['Distance_from_Home'].value_counts(dropna=False).plot(kind='bar')
plt.title('Distance_from_Home')
plt.show()

print(student_performance['Distance_from_Home'].value_counts(dropna=False))
```

## Rellenar los NaN

```python
moda_TQ = student_performance['Teacher_Quality'].mode()[0]
student_performance['Teacher_Quality'] = student_performance['Teacher_Quality'].fillna(moda_TQ)

print(student_performance['Teacher_Quality'].value_counts(dropna=False))
```

```python
moda_PEL = student_performance['Parental_Education_Level'].mode()[0]
student_performance['Parental_Education_Level'] = student_performance['Parental_Education_Level'].fillna(moda_PEL)

print(student_performance['Parental_Education_Level'].value_counts(dropna=False))
```

```python
moda_DFH = student_performance['Distance_from_Home'].mode()[0]
student_performance['Distance_from_Home'] = student_performance['Distance_from_Home'].fillna(moda_DFH)

print(student_performance['Distance_from_Home'].value_counts(dropna=False))
```
