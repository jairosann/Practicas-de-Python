### Ejercicio 1. Restaurante
Crea una clase llamada Restaurant. El método __init__() para Restaurant debería almacenar dos atributos: restaurant_name y cuisine_type (nombre del restaurante y tipo de cocina, respectivamente). Crea un método llamado describe_restaurant() que imprima estas dos piezas de información, y un método llamado open_restaurant() que imprima un mensaje indicando que el restaurante está abierto.<br>
Crea una instancia llamada restaurant a partir de tu clase. Imprime los dos atributos individualmente, y luego llama a ambos métodos.<br>



```python
class Restaurant:
    def __init__(self, restaurant_name, cuisine_type):
        # Inicializa el nombre y el tipo de cocina del restaurante
        self.restaurant_name = restaurant_name
        self.cuisine_type = cuisine_type

    def describe_restaurant(self):
        # Muestra la información básica del restaurante
        print(f"Nombre: {self.restaurant_name}")
        print(f"Tipo de cocina: {self.cuisine_type}")

    def open_restaurant(self):
        # Informa de que el restaurante está abierto
        print(f"El restaurante {self.restaurant_name} está abierto al público.")
        

# Se crea una instancia con nombre y tipo de cocina
restaurante1 = Restaurant('Estoque', 'Española')

# Se muestran los datos y el estado del restaurante
restaurante1.describe_restaurant()
restaurante1.open_restaurant()

```

### Ejercicio 3. Usuarios
Crea una clase llamada User. Crea dos atributos llamados first_name y last_name (nombre y apellido), y luego crea varios otros atributos que normalmente se almacenan en un perfil de usuario. Crea un método llamado describe_user() que imprima un resumen de la información del usuario. Crea otro método llamado greet_user() que imprima un saludo personalizado al usuario.<br>
Crea varias instancias representando diferentes usuarios, y llama a ambos métodos para cada usuario.

```python
class User:
    def __init__(self, first_name, last_name, dni, password):
        # Inicializa los datos básicos del usuario
        self.first_name = first_name
        self.last_name = last_name
        self.dni = dni
        self.password = password

    def describe_user(self):
        # Muestra un resumen con la información del usuario
        print(f"Nombre: {self.first_name}")
        print(f"Apellido: {self.last_name}")
        print(f"DNI: {self.dni}")
        print(f"Contraseña: {self.password}")

    def greet_user(self):
        # Saludo personalizado al usuario
        print(f"Hola, {self.first_name}. Bienvenido de nuevo.")


# Creación de varios usuarios de ejemplo
usuario1 = User('Jairo', 'Sánchez', 53910592, 'Jairosan')
usuario2 = User('María', 'López', 72849302, 'MaryLpz')
usuario3 = User('Carlos', 'Ruiz', 64582019, 'Cruz19')

# Llamadas a los métodos para cada usuario
usuario1.describe_user()
usuario1.greet_user()

usuario2.describe_user()
usuario2.greet_user()

usuario3.describe_user()
usuario3.greet_user()

```

### Ejercicio 4. Número de Clientes Servidos
Empieza con tu programa del Ejercicio 1. Añade un atributo llamado `number_served` con un valor por defecto de 0. Crea una instancia llamada `restaurant` a partir de esta clase. Imprime el número de clientes que el restaurante ha servido, y luego cambia este valor e imprímelo de nuevo.<br>
Añade un método llamado `set_number_served()` que te permita establecer el número de clientes que han sido servidos. Llama a este método con un nuevo número e imprime el valor de nuevo.<br>
Añade un método llamado `increment_number_served()` que te permita incrementar el número de clientes que han sido servidos. Llama a este método con cualquier número que pueda representar cuántos clientes fueron servidos, por ejemplo, en un día de negocios.



```python
class Restaurant:
    def __init__(self, restaurant_name, cuisine_type):
        # Inicializa el nombre, el tipo de cocina y los clientes servidos
        self.restaurant_name = restaurant_name
        self.cuisine_type = cuisine_type
        self.number_served = 0  # valor inicial por defecto

    def describe_restaurant(self):
        # Muestra la información básica del restaurante
        print(f"{self.restaurant_name} ofrece cocina {self.cuisine_type}.")

    def open_restaurant(self):
        # Indica que el restaurante está abierto
        print(f"El restaurante {self.restaurant_name} está abierto.")

    def set_number_served(self, number):
        # Establece el número total de clientes servidos
        self.number_served = number

    def increment_number_served(self, additional):
        # Suma nuevos clientes al total
        self.number_served += additional


# Creación de la instancia
restaurant = Restaurant("Estoque", "Española")

# Muestra el número inicial de clientes
print(f"Clientes servidos: {restaurant.number_served}")

# Actualiza el número directamente
restaurant.number_served = 20
print(f"Clientes servidos (actualizado directamente): {restaurant.number_served}")

# Actualiza el número usando el método correspondiente
restaurant.set_number_served(35)
print(f"Clientes servidos (con set_number_served): {restaurant.number_served}")

# Incrementa el total simulando un día de trabajo
restaurant.increment_number_served(10)
print(f"Clientes servidos (tras incremento): {restaurant.number_served}")
```

### Ejercicio 5. Intentos de Inicio de Sesión
Añade un atributo llamado `login_attempts` a tu clase `User` del Ejercicio 3. Escribe un método llamado `increment_login_attempts()` que incremente el valor de `login_attempts` en 1. Escribe otro método llamado `reset_login_attempts()` que reinicie el valor de `login_attempts` a 0.<br>
Crea una instancia de la clase `User` y llama a `increment_login_attempts()` varias veces. Imprime el valor de `login_attempts` para asegurarte de que se ha incrementado correctamente, y luego llama a `reset_login_attempts()`. Imprime `login_attempts` de nuevo para asegurarte de que se ha reiniciado a 0.

```python
class User:
    def __init__(self, first_name, last_name, dni, password):
        # Inicializa los datos básicos y los intentos de inicio de sesión
        self.first_name = first_name
        self.last_name = last_name
        self.dni = dni
        self.password = password
        self.login_attempts = 0  # valor inicial

    def describe_user(self):
        # Muestra los datos del usuario
        print(f"Nombre: {self.first_name} {self.last_name}")
        print(f"DNI: {self.dni}")
        print(f"Contraseña: {self.password}")

    def greet_user(self):
        # Saludo personalizado
        print(f"Hola, {self.first_name}. Bienvenido de nuevo.")

    def increment_login_attempts(self):
        # Suma un intento de inicio
        self.login_attempts += 1

    def reset_login_attempts(self):
        # Restablece el contador a cero
        self.login_attempts = 0


# Creación de un usuario de ejemplo
usuario1 = User('Jairo', 'Sánchez', 53910592, 'Jairosan')

# Simulación de varios intentos de inicio
usuario1.increment_login_attempts()
usuario1.increment_login_attempts()
usuario1.increment_login_attempts()

print(f"Intentos de inicio de sesión: {usuario1.login_attempts}")

# Reinicio del contador de intentos
usuario1.reset_login_attempts()
print(f"Intentos de inicio de sesión tras reinicio: {usuario1.login_attempts}")
```
