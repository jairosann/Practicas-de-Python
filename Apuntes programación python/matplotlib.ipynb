### Dos formas de crear gráficos

Hay dos formas principales de crear gráficos en matplotlib.

1. `matplotlib.pyplot.plot()` - Recomendado para gráficos simples (por ejemplo, x e y).
2. `matplotlib.pyplot.XX` (donde XX puede ser uno de muchos métodos, esto se conoce como la API orientada a objetos) - Recomendado para gráficos más complejos (por ejemplo, [`plt.subplots()`](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.subplots.html#matplotlib.pyplot.subplots) para crear múltiples gráficos en la misma figura; hablaremos de esto más adelante).

Ambos métodos suelen construirse partiendo de `import matplotlib.pyplot as plt` como base.

### 1. Primer ejemplo

Importamos la librería *NumPy* para crear datos

```python
import numpy as np
```

Ahora vamos a crear datos con *NumPy* que luego representaremos en una gráfica. La variable $x$ contiene un rango de valores entre $0$ y $2\pi$. La variable $y$ contiene el seno de $x$.   

```python
x = np.arange(0, 2*np.pi, 0.01)
y = np.sin(x)
```

Y por último representamos los datos anteriores en una gráfica. En la siguiente celda se realizan las siguientes tareas:

- Se crea la figura en la que se va a dibujar el gráfico con la llamada a <a href="http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.figure">plt.figure</a>. Se puede especificar el tamaño con el argumento *figsize*.

- Se crea el gráfico con la llamada a <a href="http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.plot">plt.plot</a>. Hay que especificar los valores *x* e *y*, que deberán ser arrays o similares de igual longitud, y opcionalmente una cadena que indica el estilo del gráfico. En el ejemplo se usa '-r', '-' para indicar que se quiere una línea continua y 'r' para indicar color rojo.

- Se dibuja una cuadrícula con <a href="http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.grid">plt.grid</a>, se crea un título para el gráfico con <a href="http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.title">plt.title</a> y se rotulan los ejes con <a href="http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.xlabel">plt.xlabel</a> y <a href="http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.ylabel">plt.ylabel</a>.


```python
# Dibujamos gráfica
plt.figure(figsize=(1, 0.5))
plt.plot(x, y, '-r')
plt.grid()
plt.title('funcion seno')
plt.xlabel('x')
plt.ylabel('sin x')
plt.show()
```

### 2. Gráficos de barras

Para dibujar gráficos de barras utilizamos la función <a href="http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.bar">plt.bar</a>. Es necesario pasarle como argumentos los valores *x* e *y*, y opcionalmente etiquetas para las barras, el alineamiento de estas con respecto a los valores de *x*, el color de las barras, etc. 

A continuación se muestran algunos ejemplos.

```python
x = np.array(range(8))
y = np.random.randint(0, 100, 8)

print(x)
print(y)
```

```python
plt.figure(figsize=(8, 4))
plt.bar(x, y)
plt.title('Barras')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.bar(x, y, align='center', tick_label=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])
plt.title('Barras')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

Es posible también dibujar gráficos de barras horizontales utilizando la función <a href="http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.barh">plt.barh</a>.

```python
plt.figure(figsize=(8, 4))
plt.barh(x, y, align='center', tick_label=['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])
plt.title('Barras')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 3. Histogramas

La función <a href="http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.hist">plt.hist</a> dibuja histogramas. Sus argumentos principales son un array o secuencia que contiene los datos, y el número de cajas o *bins* (por defecto igual a 10). La función divide el rango de valores de los datos en un conjunto de cajas equiespaciadas, cuenta el número de puntos que caen en cada caja y dibuja el resultado usando barras verticales. 

A continuación se muestra un ejemplo.

```python
# Conjunto de datos de 10000 distribuidos segun una normal N(1, 0)
x = np.random.random(10000)
```

```python
plt.figure(figsize=(8, 4))
plt.hist(x, bins=50)
plt.title('Histograma')
plt.xlabel('x')
plt.ylabel('frecuencia')
plt.show()
```

### 4. Gráficos de dispersión

La función <a href="http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.scatter">plt.scatter</a> se utiliza para representar diagramas de dispersión de dos variables (*x* e *y*). Como argumentos opcionales se puede especificar el color de los puntos o el tipo y tamaño del marcador.

A continuación se muestran dos ejemplos.
 

```python
plt.figure(figsize=(10, 6))
plt.scatter(x, y)
plt.title('Dispersion')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

### 6. Varios gráficos en la misma figura

Es posible dividir una figura en varios paneles, y dibujar un gráfico en cada uno de ellos. Para conseguirlo se utiliza la función <a href="http://matplotlib.org/api/pyplot_api.html#matplotlib.pyplot.subplot">plt.subplot</a>, con tres argumentos que indican el número de filas, el número de columnas y la posición en la que se quiere dibujar el siguiente gráfico. La posición empieza en 1 y se cuenta por filas, de izquierda a derecha.

Por ejemplo la llamada *plt.subplot(2, 2, 3)* dividirá la figura en 4 subfiguras (2 filas y 2 columnas) y hará que el siguiente gráfico se dibuje en la subfigura correspondiente a la fila 2 y la columna 1.
