Ejercicio completo

Crea una clase ‘CalculadoraAvanzada’ que realice operaciones matemáticas básicas (suma, resta, multiplicación y división) y permita acceder a resultados históricos. Implementa los siguientes requisitos:

1. La clase debe tener un diccionario para almacenar los resultados de las operaciones con las claves "suma", "resta", "multiplicación" y "división".

2. Implementa el método ‘__init__’ para inicializar el diccionario.

3. Implementa un método ‘operar’ que realice una operación matemática basada en un operador dado (‘+’, ‘-’, ‘*’, ‘/’).

4. Implementa un método ‘historial’ que devuelva el resultado de una operación dado el índice de la operación que se quiere obtener. Es decir, el índice indica la posición del resultado a devolver. Por ejemplo, si he realizado y almacenado tres sumas y cinco divisiones, podría llamar al método 'historial' para que me devuelva el resultado de la suma número 2 y luego hacer otra llamada que devuelva el resultado de la división número 3.

5. Implementa dos excepciones:

  - Una para manejar la división por cero.

  - Otra para manejar problemas de índice al acceder al historial.

6. Cree una instancia de la clase en el programa principal, realice unas cuantas operaciones e imprima el historial de alguna de ellas.
