## Parte 1

Crea las siguientes clases:

- `Planta`: con un atributo `nombre` y un método `crecer()` que imprima un mensaje como `"La planta [nombre] está creciendo."`
- `Animal`: con un atributo `nombre` y un método `comer()` que imprima `"El animal [nombre] está buscando comida."`
- `Ambiente`: con un atributo `nombre` y con listas de `animales` y `plantas`.
  - Un método `mostrar_ecosistema()` que imprima qué seres hay en el ambiente.
  - Los métodos para añadir plantas o animales al ambiente.

**Nota:** El ambiente debe “contener” animales y plantas.


````python
class Planta:

    def __init__(self, nombre):
        self.nombre = nombre


    def crecer(self):
        print(f"La planta {self.nombre} está creciendo.")


class Animal:
    def __init__(self, nombre):
        self.nombre = nombre

    def comer(self):
        print(f"El animal {self.nombre} está buscando comida.")

class Ambiente:
    def __init__(self, nombre):
        self.nombre = nombre
        self.animales = []
        self.plantas = []

    def añadir_planta(self, planta):
        self.plantas.append(planta)

    def añadir_animal(self, animal):
        self.animales.append(animal)

    def mostrar_ecosistema(self):
        print(f"Ambiente: {self.nombre}")
        print("Plantas:")
        for p in self.plantas:
            print(f" - {p.nombre}")
        print("Animales:")
        for a in self.animales:
            print(f" - {a.nombre}")


# Crear plantas
p1 = Planta("Helecho")
p2 = Planta("Margarita")

# Crear animales
a1 = Animal("Lobo")
a2 = Animal("Zorro")

# Crear ambiente
bosque = Ambiente("Bosque")

# Añadir seres al ambiente
bosque.añadir_planta(p1)
bosque.añadir_planta(p2)
bosque.añadir_animal(a1)
bosque.añadir_animal(a2)

# Probar métodos
p1.crecer()
a1.comer()
bosque.mostrar_ecosistema()
````
